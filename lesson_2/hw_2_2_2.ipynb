Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.

```python
import pandas as pd

authors = pd.DataFrame({'author_id':[1, 2, 3],
                        'author_name':['Тургенев', 'Чехов', 'Островский']}, 
                        columns = ['author_id', 'author_name'])

books = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],
                     'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 
                                   'Толстый и тонкий', 'Дама с собачкой', 
                                   'Гроза', 'Таланты и поклонники'],
                     'price':[450, 300, 350, 500, 450, 370, 290]}, 
                     columns = ['author_id', 'book_title', 'price'])
```

```python
authors
```

```python
books
```

```python
authors_price = pd.merge(authors, books, on = 'author_id', how = 'inner')

authors_price
```

```python
authors_price = pd.merge(authors, books, on = 'author_id', how = 'left')

authors_price
```

```python
authors_price = pd.merge(authors, books, on = 'author_id', how = 'right')

authors_price
```

```python
authors_price = pd.merge(authors, books, on = 'author_id', how = 'outer')

authors_price
```

```python
authors_mean_price = pd.merge(authors, books, on = 'author_id', how = 'inner')

authors_mean_price = authors_mean_price.drop('book_title', axis=1)

authors_mean_price = authors_mean_price.drop('author_id', axis=1)

authors_mean_price
```

```python
authors_mean_price = authors_mean_price.groupby('author_name').agg({'price':'mean'})

authors_mean_price
```

```python
authors_mean_price = authors_mean_price.reset_index()

authors_mean_price = authors_mean_price.rename(columns={'price':'mean_price'})

authors_mean_price
```
